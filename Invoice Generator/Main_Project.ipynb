```python
def my_temp(c):
    
    
    #Setting some margin
    c.translate(inch,inch)

    #drawing one line
    c.setStrokeColor("Red")
    c.setLineWidth(5)
    c.line(0,8*inch,7*inch,8*inch)  #line(x1,y1,x2,y2) , x1 and y1 are the starting coordinates and x2,y2 ending

    #Increasing th font size and changing the font of the text(Data String)
    c.setFillColor("Blue")
    c.setFont("Helvetica",20)

    # Writing in the pdf
    #c.drawString(200,200,"Hello World")

    #Adding an image
    c.drawImage('C:\\Users\\Hp\\Invoice Generator\\logo2.jpg',-0.8 * inch,8.1*inch)

    #Writing invoice geneartion
    c.setFillColor("Black")
    c.setFont("Helvetica",25)
    c.drawString(3*inch,8.4*inch,"Invoice")

    #Drawing a rectangle
    #c.setLineWidth(3)
    #c.setStrokeColor('black')
    #c.setFillColor('white')
    #c.rect(1*inch,2*inch,3.5*inch,5*inch,fill =1)   #c.rect(x1,y1,x2,y2) , x1,y1 are the coordinates of the 
    #c.rect(0,0,6.5*inch,7*inch,fill =1)
    #left bottom corner of rectangle and x2,y2 are the coordinates of right upper corner of rectangle


    #Adding the address and time
    from datetime import datetime
    # datetime object containing current date and time
    now = datetime.now()
    # dd/mm/YY H:M:S
    dt_string = now.strftime("%d/%m/%Y %H:%M:%S")
    print("date and time =", dt_string)

    my_style = ParagraphStyle("My Para Style",fontSize = 12)
    width,height = A4
    p1 = Paragraph('''<b>Service options </b>: In-store shopping · In-store pick-up · Delivery <BR/>
    <b> Address: </b> M75M+3P9, Vijay Chowk, Block F, Krishna Nagar, Delhi, 110051 <BR/>
    <b> Contact: </b> 98320XXXXX
    <BR/>
    <b> Invoice Date and time: {time}</b> '''.format(time = dt_string),my_style)
    p1.wrapOn(c,200,50) #size of the paragraph
    p1.drawOn(c,width-270,height-250) #where the paragraph will be placed

    
    c.setFont("Times-Roman",22)
    c.setFillColorRGB(0,0,0)
    c.setStrokeColor("Black")
    c.setLineWidth(2)
    c.drawString(0.5*inch,7.3*inch,"Products")
    c.drawString(2*inch,7.3*inch,"Qty")
    c.drawString(3.5*inch,7.3*inch,"Price")
    c.drawString(5*inch,7.3*inch,"Discount")
    c.drawString(6.3*inch,7.3*inch,"Amount")
    
    
    #vertical line color
    c.line(1.7*inch,7.5*inch,1.7*inch,2*inch)
    c.line(3.2*inch,7.5*inch,3.2*inch,2*inch)
    c.line(4.7*inch,7.5*inch,4.7*inch,2*inch)
    c.line(6.2*inch,7.5*inch,6.2*inch,2*inch)
    
    #drawing line
    #drawing one line
    c.setStrokeColor("Black")
    c.setLineWidth(2)
    c.line(0,7*inch,7*inch,7*inch)
    c.line(0,2*inch,7*inch,2*inch)
    
    #Taxes
    c.drawString(0,1.5*inch,"Total Taxable charges :-")
    c.drawString(0,1*inch,"Total Taxes :-")
    c.drawString(0,0.5*inch,"Total Value :-")
    c.drawString(5*inch,0,"Signature")
    
    return c


from reportlab.pdfgen import canvas
from reportlab.lib.units import inch
from reportlab.lib.pagesizes import A4,letter
from reportlab.platypus import Paragraph
from reportlab.lib.styles import ParagraphStyle



my_path = 'C:\\Users\\Hp\\Invoice Generator\\Invoice.pdf'   #file Path
c = canvas.Canvas(my_path,pagesize= letter)
c = my_temp(c)

# Taking input from the user
DATA = []
HEAD = ['item name','Quantity','Unit Price','Discount','Total Price']

DATA.append(HEAD)

#Item info
n_items = int(input("Enter the number of items :"))
for i in range(n_items): #this loop gonna run for n_items no of items
        name = input("Enter item name :")
        quant = int(input("Enter the quantity of item:"))
        unitP = float(input('Enter unit price of the item'))
        discount = float(input("Enter the percentage of discount"))
        totalP1 = quant*unitP
        totalP2 = totalP1 - (discount*(totalP1))/100
        DATA.append([name,quant,unitP,discount,totalP2])
        
#print(DATA)

#Footer
FOOT = ['','','Total amount = ',0]
#list slicing
for i in DATA[1:]:
    FOOT[3] = FOOT[3] + i[4]
    #print(i[3])
     
DATA.append(FOOT)
print(DATA)
total = FOOT[3]
print(type(total))
print(total)

c.setFillColorRGB(0,0,1)
c.setFont("Helvetica",20)
row_gap  = 0.6
line_y = 6.5

for i in range(n_items):
    c.drawString(0.1*inch,line_y*inch,str(DATA[i+1][0]))
    c.drawString(2*inch,line_y*inch,str(DATA[i+1][1]))
    c.drawString(3.5*inch,line_y*inch,str(DATA[i+1][2]))
    c.drawString(5*inch,line_y*inch,str(DATA[i+1][3]))
    c.drawString(6.3*inch,line_y*inch,str(DATA[i+1][4]))
    line_y = line_y - row_gap

c.drawString(3*inch,1.5*inch,str(total))
totalTaxesP = float(input("Enter the percentage of taxes"))
totalTaxes = (total*totalTaxesP)/100
payable = total + totalTaxes
c.drawString(2*inch,1*inch,str(totalTaxes))
c.drawString(2*inch,0.5*inch,str(payable))

c.showPage() #
c.save() #Save the pdf
```
